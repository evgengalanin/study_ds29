### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.



```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.tail()
```

### Фильтруем данные

```python
given_level = 5.0
```

```python
given_level_ratings = ratings.query(f'rating == {given_level}')
given_level_ratings.head()
```

```python
movie_counter = given_level_ratings['movieId'].value_counts().head()
movie_counter
```

```python
top_movie = movies.query(f'movieId == {movie_counter.index[0]}')
top_movie
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.



```python
data = pd.read_csv('power.csv')
data.head()
```

### Задаём переменные

```python
data[ data['country'].str.contains('est', case=False) ]['country'].unique()
```

```python
baltic = ['Lithuania', 'Latvia', 'Estonia']
```

```python
category = [4, 12, 21]
```

```python
start_date = 2005
end_date = 2010
```

```python
period = list(range( start_date, end_date + 1 ))
period
```

### Фильтруем данные

```python
filtered_baltic_countries = data.query(f'country in { baltic } and year in { period } and category in { category } and quantity > 0')
filtered_baltic_countries
```

```python
filtered_baltic_countries['quantity'].sum()
```
