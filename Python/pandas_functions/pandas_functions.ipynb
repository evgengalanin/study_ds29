### Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
def count_mean_rating0(data):
    '''Подсчёт среднего значения оценки фильма и присвоение рейтинга'''
    if data.rating.mean() < 2:
        return 'Низкий рейтинг'
    elif data.rating.mean() < 4:
        return 'Средний рейтинг'
    else:
        return 'Высокий рейтинг'
```

```python
mean_ratings = ratings.groupby('movieId').apply( count_mean_rating0 )
```

```python
def classify_movies(row):
    '''Классификация фильмов по рейтингу'''
    try:
        return mean_ratings.loc[row['movieId']]
    except:
        return 'Nan'
```

```python
movies['class'] = movies.apply(classify_movies, axis=1)
movies
```

### Задание 2

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {
            'Центр': ['москва', 'тула', 'ярославль'],
            
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
            }

Результат классификации запишите в отдельный столбец region.

```python
key_words = pd.read_csv('keywords.csv')
key_words
```

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
```

```python
def asign_geoloc_region(row):
    """Присвоение географического региона каждому запросу внутри строки row"""
    
    for key, values in geo_data.items():
        for value in values:
            if value in row['keyword']:
                return key
            else:
                return 'Undefined'
```

```python
key_words['region'] = key_words.apply(asign_geoloc_region, axis=1)
key_words
```

для каждой строки пройдите по всем годам списка years
если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска

### 1. В датафрейме movies в столбец year выставляем год выпуска для каждого фильма

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
years = []

for year in range(1950, 2010+1):
    years.append(str(year))
```

```python
def prodaction_year(row):
    """Выставляет каждой строке год выпуска фильма"""
    
    for year in years:
        if year in row['title']:
            return year
        
    return '1900'   
```

```python
movies['year'] = movies.apply(prodaction_year, axis=1)
movies
```

### 2. Используя датафрейм ratings считаем среднюю оценку для каждого фильма и заносим в movies в столбец rating

```python
def count_mean_rating(data):
    return data.rating.mean()
```

```python
mean_ratings_movies = ratings.groupby('movieId').apply( count_mean_rating )
```

```python
def get_average_rating(row):
    try:
        return mean_ratings_movies.loc[row['movieId']]
    except:
        return 0.0
```

```python
movies['rating'] = movies.apply(get_average_rating, axis=1)
movies
```

### 3. Считаем среднюю оценку для каждого года

```python
mean_ratings_years = movies.groupby('year').apply( count_mean_rating )
```
